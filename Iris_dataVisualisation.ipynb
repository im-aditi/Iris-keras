```python
from sklearn.datasets import load_iris
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris=load_iris()
x=iris.data
y=iris.target
x=x.astype('float32')
print(x.shape,y.shape,sep='\n')
```

```python
data=pd.DataFrame(data=x,columns=['Sepal Length','Sepal Width','Petal Length','Petal Width'])
target=[]
for i in y:
    if i==0:
        target.append('Setosa')
    elif i==1:
        target.append('Versicolor')
    else:
        target.append('Virginica')
data['Label']=target
```

```python
sns.pairplot(data,hue='Label',palette='husl')
```
